```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

# Load Dataset

```python
dataset = load_iris()
```

```python
x, y, label, features = dataset.data, dataset.target, dataset.target_names, dataset.feature_names
```

```python
x
```

```python
y
```

```python
label
```

# Decision Tree

```python
dtc = DecisionTreeClassifier()
dtc02 = DecisionTreeClassifier(criterion="entropy")
```

```python
model01 = dtc.fit(x, y)
model02 = dtc02.fit(x,y)
```

```python
label[model01.predict([[2.5, 3, 0.1, 4]])[0]]
```

# Plot

```python
plot_tree(model01)
```

```python
plt.figure(figsize=(15, 12))
plot_tree(model01, filled= True, class_names=label, feature_names=features)
plt.show()
```

```python
plt.figure(figsize=(15, 12))
plot_tree(model02, filled= True, class_names=label, feature_names=features)
plt.show()
```
