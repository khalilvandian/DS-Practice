```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

# Load Dataset

```python
dataset = load_iris()
```

```python
x, y, label, features = dataset.data, dataset.target, dataset.target_names, dataset.feature_names
```

```python
x
```

```python
y
```

```python
label
```

# Decision Tree

```python
dtc = DecisionTreeClassifier()
dtc02 = DecisionTreeClassifier(criterion="entropy")
pruned_dtc01 = DecisionTreeClassifier(ccp_alpha=0.001)
pruned_dtc02 = DecisionTreeClassifier(ccp_alpha=0.01)
pruned_dtc03 = DecisionTreeClassifier(ccp_alpha=0.1)
pruned_dtc04 = DecisionTreeClassifier(ccp_alpha=1)
```

```python
model01 = dtc.fit(x, y)
model02 = dtc02.fit(x,y)
model03 = pruned_dtc01.fit(x, y)
model04 = pruned_dtc02.fit(x, y)
model05 = pruned_dtc03.fit(x, y)
model06 = pruned_dtc04.fit(x, y)
```

```python
label[model01.predict([[2.5, 3, 0.1, 4]])[0]]
```

# Plot

```python
plot_tree(model01)
```

```python
plt.figure(figsize=(15, 12))
plot_tree(model01, filled= True, class_names=label, feature_names=features)
plt.show()
```
