```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

# Load Dataset

```python
dataset = load_iris()
```

```python
x, y, label = dataset.data, dataset.target, dataset.target_names
```

```python
x
```

```python
y
```

```python
label
```

# Decision Tree

```python
dtc = DecisionTreeClassifier()
```

```python
model01 = dtc.fit(x, y)
```

```python
label[model01.predict([[2.5, 3, 0.1, 4]])[0]]
```

# Plot

```python
plot_tree(model01)
```

```python
plt.figure(figsize=(15, 12))
plot_tree(model01, filled= True)
```
